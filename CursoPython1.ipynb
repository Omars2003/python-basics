# Expresiones logicas y control de flujo  üé¥

Las expresiones l√≥gicas son combinaciones de operadores l√≥gicos y operandos que eval√∫an condiciones y devuelven un valor booleano, es decir, True (verdadero) o False (falso).  

Las mas comunes:

 ‚úç (y l√≥gico): Devuelve True si ambas condiciones son verdaderas.

 ‚úçor (o l√≥gico): Devuelve True si al menos una de las condiciones es verdadera.

‚úç not (negaci√≥n l√≥gica): Invierte el valor de una condici√≥n. Si la condici√≥n es verdadera, devuelve False; si es falsa, devuelve True.

Ejemplos ‚öì

üí•  5 > 3 and 2 < 1 devuelve False porque la primera condici√≥n es verdadera, pero la segunda es falsa.

üí•5 > 3 or 2 < 1 devuelve True porque al menos una de las condiciones es verdadera.

üí•not (4 == 4) devuelve False porque la condici√≥n inicial es verdadera, pero la negaci√≥n invierte su valor.

 üë∫ Las expresiones l√≥gicas son ampliamente utilizadas en estructuras de control condicionales, como if, elif y while, para tomar decisiones basadas en condiciones y controlar el flujo de ejecuci√≥n del programa.

```python
# Solicitar el nombre y la edad del usuario
nombre = input("Ingresa tu nombre: ")
edad = int(input("Ingresa tu edad: "))

# Verificar si el usuario es mayor de edad y tiene permiso de conducir
if edad >= 18:
    tiene_permiso_conducir = input("¬øTienes permiso de conducir? (s/n): ")
    if tiene_permiso_conducir.lower() == "s":
        print("Hola,", nombre, "Eres mayor de edad y tienes permiso de conducir.")
    else:
        print("Hola,", nombre, "Eres mayor de edad, pero no tienes permiso de conducir.")
else:
    print("Hola,", nombre, "Eres menor de edad. No puedes conducir legalmente.")

numero = float(input("Ingresa un n√∫mero: "))

if numero > 0:
    print("El n√∫mero", numero, "es positivo.")
elif numero < 0:
    print("El n√∫mero", numero, "es negativo.")
else:
    print("El n√∫mero es cero.")

# Verificar si el n√∫mero es divisible por 3 y 5 simult√°neamente
if numero % 3 == 0 and numero % 5 == 0:
    print("El n√∫mero", numero, "es divisible por 3 y 5 al mismo tiempo.")
else:
    print("El n√∫mero", numero, "no es divisible por 3 y 5 al mismo tiempo.")
```
